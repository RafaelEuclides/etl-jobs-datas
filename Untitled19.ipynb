```python
import pandas as pd
import json

# Nome do arquivo JSON contendo os dados anuais
file_path = "profissoes_anuais.json"

try:
    # --- 1. Carregar o arquivo com a biblioteca 'json' ---
    # Carrega a estrutura completa do JSON
    with open(file_path, 'r', encoding='utf-8') as f:
        dados_json = json.load(f)

    # Acessa a lista principal de profiss√µes (a chave 'profissoes_anuais')
    profissoes_data = dados_json['profissoes_anuais']

    # --- 2. Cria√ß√£o, Normaliza√ß√£o e Pivotagem do DataFrame ---

    # 2a. Criar o DataFrame inicial a partir da lista.
    # A coluna 'contagem' ser√° uma coluna de dicion√°rios.
    df_profissoes = pd.DataFrame(profissoes_data)

    # 2b. Normalizar a coluna 'contagem' (expandindo os anos em colunas)
    df_contagem_expandida = pd.json_normalize(df_profissoes['contagem'])

    # 2c. Concatenar as colunas de identifica√ß√£o ('cargo', 'area') com as colunas anuais expandidas
    df_tabela_final = pd.concat([df_profissoes[['cargo', 'area']], df_contagem_expandida], axis=1)

    # 2d. Renomear e definir 'cargo' e 'area' como √≠ndice (formato de tabela)
    df_tabela_final.rename(columns={'cargo': 'Cargo', 'area': '√Årea'}, inplace=True)
    df_tabela_final = df_tabela_final.set_index(['Cargo', '√Årea'])

    # 2e. Configurar o √≠ndice das colunas (Anos)
    df_tabela_final.columns.name = 'Ano'
    df_tabela_final.columns = df_tabela_final.columns.astype(int)

    # --- 3. Exibir a Tabela Final ---

    print("Tabela de Contagem de Profissionais por Ano (2022-2025)")

    # Exibir a tabela completa (usando print() para consist√™ncia)
    print(df_tabela_final)

except FileNotFoundError:
    print(f"Erro: O arquivo '{file_path}' n√£o foi encontrado. Certifique-se de que ele est√° no mesmo diret√≥rio do seu script.")
except json.JSONDecodeError:
    print(f"Erro: O arquivo '{file_path}' n√£o √© um JSON v√°lido. Verifique a sintaxe.")
except Exception as e:
    print(f"Ocorreu um erro inesperado: {e}")
```

```python
import pandas as pd
import json

# Nome do arquivo JSON contendo os dados anuais
file_path = "profissoes_anuais.json"

try:
    # --- 1. Carregar e Normalizar o JSON ---
    with open(file_path, 'r', encoding='utf-8') as f:
        dados_json = json.load(f)

    profissoes_data = dados_json['profissoes_anuais']
    df_profissoes = pd.DataFrame(profissoes_data)
    df_contagem_expandida = pd.json_normalize(df_profissoes['contagem'])

    # Criar a tabela com 'Cargo' e '√Årea' como √≠ndice e Anos como colunas
    df_tabela_final = pd.concat([df_profissoes[['cargo', 'area']], df_contagem_expandida], axis=1)
    df_tabela_final.rename(columns={'cargo': 'Cargo', 'area': '√Årea'}, inplace=True)
    df_tabela_final = df_tabela_final.set_index(['Cargo', '√Årea'])

    # Converter colunas de ano para inteiro (necess√°rio para c√°lculo e ordena√ß√£o)
    df_tabela_final.columns = df_tabela_final.columns.astype(int)

    # --- 2. Aplicar a L√≥gica de C√°lculo de M√©dia (Seu Exemplo) ---

    # df.iloc[:, 0] √© '2022', df.iloc[:, 3] √© '2025'.
    # Iloc [:, 0:4] seleciona as colunas 2022, 2023, 2024, 2025.
    df_tabela_final['M√©dia Anual (2022-2025)'] = df_tabela_final.iloc[:, 0:4].mean(axis=1)

    # --- 3. Selecionar e Renomear Colunas (Seu Exemplo) ---

    # Selecionar as colunas '√Årea' (n√≠vel 1 do √≠ndice) e a nova coluna 'M√©dia' (√∫ltima coluna)
    # Resetamos o √≠ndice para que 'Cargo' e '√Årea' virem colunas normais
    df_transformado = df_tabela_final.reset_index()

    # Selecionamos o Cargo, √Årea e a M√©dia calculada (√∫ltima coluna)
    df_transformado = df_transformado.iloc[:, [0, 1, -1]]

    # Renomear colunas
    df_transformado = df_transformado.rename(columns={
        df_transformado.columns[0]: 'Cargo',
        df_transformado.columns[1]: '√Årea',
        df_transformado.columns[-1]: 'M√©dia de Contagem'
    })

    # --- 4. Visualizar o DataFrame Transformado ---

    print("M√©dia Anual de Profissionais por Cargo (2022-2025)")
    print(df_transformado.sort_values(by='M√©dia de Contagem', ascending=False))

except FileNotFoundError:
    print(f"Erro: O arquivo '{file_path}' n√£o foi encontrado.")
except Exception as e:
    print(f"Ocorreu um erro inesperado: {e}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import json

# Nome do arquivo JSON contendo os dados anuais
file_path = "profissoes_anuais.json"

try:
    # --- 1. Carregar e Normalizar o JSON (Para recriar o df_transformado) ---
    with open(file_path, 'r', encoding='utf-8') as f:
        dados_json = json.load(f)

    profissoes_data = dados_json['profissoes_anuais']
    df_profissoes = pd.DataFrame(profissoes_data)
    df_contagem_expandida = pd.json_normalize(df_profissoes['contagem'])

    # Criar a tabela com 'Cargo' e '√Årea' como √≠ndice
    df_tabela_final = pd.concat([df_profissoes[['cargo', 'area']], df_contagem_expandida], axis=1)
    df_tabela_final.rename(columns={'cargo': 'Cargo', 'area': '√Årea'}, inplace=True)
    df_tabela_final = df_tabela_final.set_index(['Cargo', '√Årea'])
    df_tabela_final.columns = df_tabela_final.columns.astype(int)

    # Calcular a M√©dia Anual (2022-2025)
    df_tabela_final['M√©dia Anual (2022-2025)'] = df_tabela_final.iloc[:, 0:4].mean(axis=1)

    # Selecionar e Renomear Colunas para o DataFrame transformado
    df_transformado = df_tabela_final.reset_index()
    df_transformado = df_transformado.iloc[:, [0, 1, -1]]
    df_transformado.rename(columns={
        df_transformado.columns[0]: 'Cargo',
        df_transformado.columns[1]: '√Årea',
        df_transformado.columns[-1]: 'M√©dia de Contagem'
    }, inplace=True)

    # --- 2. Exportar o DataFrame transformado (Opcional, conforme seu exemplo) ---
    # output_file_path = "media_profissoes_por_cargo.csv"
    # df_transformado.to_csv(output_file_path, index=False)
    # print(f'Dados transformados exportados para {output_file_path}')


    # --- 3. Plotar o Gr√°fico de Barras (Seu Exemplo Adaptado) ---

    # Ordenar os dados pela m√©dia para que o gr√°fico fique mais claro
    df_plot = df_transformado.sort_values(by='M√©dia de Contagem', ascending=True)

    plt.figure(figsize=(12, 7))

    # Criar o gr√°fico de barras horizontais (melhor para r√≥tulos longos de cargos)
    plt.barh(df_plot['Cargo'], df_plot['M√©dia de Contagem'], color='#1f77b4')

    # Configurar t√≠tulo e r√≥tulos
    plt.title('M√©dia Anual de Contagem de Profissionais por Cargo (2022-2025)', fontsize=16)
    plt.xlabel('M√©dia de Profissionais (em milh√µes)', fontsize=12)
    plt.ylabel('Cargo', fontsize=12)

    # Formatar o eixo X para milh√µes
    from matplotlib.ticker import FuncFormatter
    def format_millions(x, pos):
        return '%1.1fM' % (x * 1e-6)

    formatter = FuncFormatter(format_millions)
    plt.gca().xaxis.set_major_formatter(formatter)

    plt.grid(axis='x', linestyle='--', alpha=0.6)
    plt.tight_layout()
    plt.show()


except FileNotFoundError:
    print(f"Erro: O arquivo '{file_path}' n√£o foi encontrado. Certifique-se de que ele est√° no mesmo diret√≥rio do seu script.")
except Exception as e:
    print(f"Ocorreu um erro inesperado: {e}")
```
